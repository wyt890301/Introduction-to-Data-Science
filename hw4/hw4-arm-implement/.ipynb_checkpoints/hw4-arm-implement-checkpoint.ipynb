### Creating APRIORI function to generate frequent itesets based on minimum threshold support = 0.04

```python
def APRIORI(data, min_support = 0.04,  max_length = 4):
    # Collecting Required Library
    import numpy as np
    import pandas as pd
    import itertools
    
    support = {} 
    L = list(data.columns)  # 所有商品
    total_count = len(data)
    
    # write your code here !!
    # 先找出frequent item
    for item in L:
        item_support = len(data[data[item]==1]) / total_count
        if item_support >= 0.04:
            support[item] = item_support

    for i in range(2, max_length+1):
        # 收集frequent item
        itemsets = []
        for frequent_item in support:
            if i > 2 and type(frequent_item) == tuple:
                for item in frequent_item:
                    if item not in itemsets:
                        itemsets.append(item)
            else:
                itemsets.append(frequent_item)
                
        # 找出所有可能的組合
        itemsets = itertools.combinations(itemsets, i)
        
        for itemset in itemsets:
            # 找出符合itemset的交易
            tempdata = data
            for item in itemset:
                tempdata = tempdata[tempdata[item]==1]
                
            # 計算此itemset之support
            itemset_support = len(tempdata)/ total_count
            if itemset_support >= 0.04:
                support[itemset] = itemset_support

    result = pd.DataFrame(list(support.items()), columns = ["Items", "Support"])
    return result
```

```python
## finding frequent itemset with min support = 4%
my_freq_itemset = APRIORI(transaction, 0.04, 3)
my_freq_itemset.sort_values(by = 'Support', ascending = False)
```

### Creating ASSOCIATION_RULE function to generate rules based on minimun threshold confidence.

```python
def ASSOCIATION_RULE(df, min_confidence_threshold = 0.5):
    import pandas as pd
    import itertools
    
    support = pd.Series(df.Support.values, index=df.Items).to_dict()
    data = []
    L = df.Items.values
    total_count = len(data)

    # write your code here !!
    for itemset in L:
        if type(itemset) == tuple:
            total_support = support[itemset]
            itemset_list = itertools.permutations(itemset)
            for subset in itemset_list:
                ant = subset[0]
                confidence = total_support/support[ant]
                if confidence > min_confidence_threshold:
                    con = subset[1]
                    ant_support = support[ant]
                    con_support = support[con]
                    lift = confidence/con_support
                    leverage = total_support - ant_support * con_support
                    convection = (1-con_support)/(1-confidence)
                    data.append([ant, con, ant_support, con_support, total_support, confidence, lift, leverage, convection])

    result = pd.DataFrame(data, columns = ["antecedents", "consequents", "antecedent support", "consequent support",
                                        "support", "confidence", "Lift", "Leverage", "Convection"])
    return result
```

```python
## Rule with minimun confidence = 50%
my_rule = ASSOCIATION_RULE(my_freq_itemset, 0.5)
my_rule
# the column names in the following table is the same as those generated by mlxtend
# you can refer to http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/association_rules/ 
# to find the formulas of lift, leverage, convection
```

### Finally sorting results by Lift to get highly associated itemsets.

## Cross Verifying results with  apriori and association rule from mlxtend

```python
# Loading standard package
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
## finding frequent itemset with min support = 4%
frequent_itemset = apriori(df = transaction, min_support= 0.04, use_colnames= True)
frequent_itemset.sort_values(by = 'support', ascending = False)
```

### Createing associate rule such that item brought with conditional probability(Confidence) more than 50% with corresponding item

```python
## Rule with minimun confidence = 50%
Rules = association_rules(frequent_itemset, min_threshold= 0.5)
Rules
[['Cake', 'Coffee', 0.10313713146574337, 0.4750813136082258, 0.5269582909460834, 0.11439929328621908, 0.0053504426317400045, 0.5550870964162876], 
 ['Pastry', 'Coffee', 0.08551043961808834, 0.4750813136082258, 0.5521472392638037, 0.09938162544169611, 0.006589941182383381, 0.550930514260544]]
```

```python
## Finally sorting results by Lift to get highly associated itemsets.
Rules.sort_values(by='lift', ascending= False).head(10)
```
